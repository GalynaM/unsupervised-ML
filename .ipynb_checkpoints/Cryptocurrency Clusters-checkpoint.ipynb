```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(Path('Resources/crypto_data.csv'), index_col = 0)
df
```

## Data Preparation

```python
# Discard all cryptocurrencies that are not being traded.
df_trade = df.loc[df['IsTrading']==True]
df_trade
```

```python
# drop the IsTrading column from the dataframe
df_col = df_trade.drop(['IsTrading'], axis = 1)
df_col
```

```python
# Remove all rows that have at least one null value
df_clean = df_col.dropna()
df_clean
```

```python
# Filter for cryptocurrencies that have been mined.
# The total coins mined should be greater than zero
df_mined = df_clean.loc[df_clean['TotalCoinsMined'] > 0]
df_mined
```

```python
# delete the CoinName from the original dataframe
df_num = df_mined.drop('CoinName', axis = 1)
df_num
```

```python
# Drop index column
df_num.reset_index(drop=True, inplace=True)
df_num
```

```python
# check column types
df_num.dtypes
```

```python
# convert TotalCoinSupply into numeric data
df_num['TotalCoinSupply'] = pd.to_numeric(df_num['TotalCoinSupply'])
df_num
```

```python
df_num.dtypes
```

```python
# convert features with text values - Algorithm and ProofType, into numerical data
X_dummies = pd.get_dummies(df_num)
X_dummies
```

```python
# Standardize dataset so that columns that contain larger values do not unduly influence the outcome.
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(X_dummies)
crypto_scaled
```

## Dimensionality Reduction

```python
# Perform dimensionality reduction with PCA
# preserve 90% of the explained variance in dimensionality reduction
pca = PCA(n_components=0.9)

crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
# Transform PCA data to a DataFrame
df_crypto_pca = pd.DataFrame(
    data=crypto_pca
)
df_crypto_pca
```

```python
# reduce the dataset dimensions with t-SNE

# Initialize t-SNE model
tsne = TSNE(learning_rate=35)
```

```python
# run t-SNE on the principal components
tsne_features = tsne.fit_transform(df_crypto_pca)
tsne_features.shape
```

```python
tsne_features
```

```python
# create a scatter plot of the t-SNE output
plt.scatter(x = tsne_features[:,0], y = tsne_features[:,1], alpha=0.5)

plt.show()
```

### Observation
* It is hard to visually detect distinct clusters.

## Cluster Analysis with k-Means

```python
# transform t-sna results to dataframe
df_crypto_tsna = pd.DataFrame(
    data=tsne_features
)
df_crypto_tsna
```

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=3)
    km.fit(df_crypto_tsna)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

### According to Elbow Curve, cryptocurrencies can be clustered by 4 or 5 groups

### Visualize the clusters

```python
# Initializing model with K = 5
model = KMeans(n_clusters=5, random_state=5)
```

```python
model.fit(df_crypto_tsna)
```

```python
# Get predictions
predictions = model.predict(df_crypto_tsna)
print(predictions)
```

```python
# Add a new class column to df_crypto_tsna
df_crypto_tsna["class"] = model.labels_
df_crypto_tsna.head()
```

```python
plt.scatter(x=df_crypto_tsna[0], y=df_crypto_tsna[1], c=df_crypto_tsna['class'])

plt.show()
```

### Conclusion
* Our investigation shows that cryptocurrencies that are on the market can be grouped to create a classification system.
* The optimal number of groups the cryptocurrencies can be clustered in varies from 4 to 5.
